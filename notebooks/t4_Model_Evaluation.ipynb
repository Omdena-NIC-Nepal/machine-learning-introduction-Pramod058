#### Evaluate the model using metrics such as Mean Squared Error (MSE), Mean Asbsolute Error (MAE), and R-squared.


##### Mean Squared Error

#### MSE: Average Squared distance between the predicted and actual values. Lower is better

```python
mse = em.meanSquaredError(y_test, y_pred)
```

```python
print("Mean squared Error:", mse)
```

#### MAE: Average absolute distance between the predicted and actual values. Lower is better

##### Mean Absoulute Error

```python
mae = em.meanAbsoluteError(y_test, y_pred)
```

```python
print("Mean Absolute Error:", mae)
```

#### R2: Variance of the dependent variable (target) explained by the indepentdent variables (features). How well the model fits in data. The higher the better

##### R-squared

```python
r2 = em.r2Score(y_test, y_pred)
```

```python
print("R2 Score:", r2)
```

#### Plot the predicted vs Actual values
- Predicted values of the target variable are plotted on the x-axis and actual values are plotted on the y-axis
- If the model is a good fit, the points should be close to the diagonal line indicating the strong relation between two values

```python
em.scatterplot(y, y_test, y_pred, alpha = 0.5)
```

#### Plot residuals to check the assumptions of linear regression.


- Residuals are plotted against the predicted values of the target variable.
- If the linear regression model is good fit for data, plot should show a random scatter of the points around 0.

```python
x = y_pred
y = (y_test.to_numpy()-y_pred.to_numpy())
alpha = 0.5
 
em.residualplot(x, y, alpha)
```
