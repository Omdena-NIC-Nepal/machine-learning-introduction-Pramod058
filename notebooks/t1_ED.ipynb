### Load the data set (Loaded from data/boston_housing_init.csv)

```python
df = pd.read_csv('../data/boston_house_prices.csv')

print(df.columns)

df.head(10)
```

##### Observation withdrawn:
- we have 15 columns that includes 'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT', 'MEDV', 'TARGET'. Where: 

- CRIM: per capita crime rate by town.
- ZN: proportion of residential land zoned for lots over 25,000 sq. ft.
- INDUS: proportion of non-retail business acres per town.
- CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise).
- NOX: nitrogen oxides concentration (parts per 10 million).
- RM: average number of rooms per dwelling.
- AGE: proportion of owner-occupied units built prior to 1940.
- DIS: weighted distances to five Boston employment centers.
- RAD: index of accessibility to radial highways.
- TAX: full-value property tax rate per $10,000.
- PTRATIO: pupil-teacher ratio by town.
- B: 1000(Bk - 0.63)^2 where Bk is the proportion of black residents by town.
- LSTAT: percentage of lower status of the population.
- MEDV: median value of owner-occupied homes in $1000s (Target).

#### Visualization of relationships between features and the target variable. 


```python
# Correlation heatmap to find how strongly each numerical feature correlates with eachother

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature correlation heatmap")
plt.show()
```

#### Some common insights: 
#### Positive corelation 
- The strong corelation of 0.91 between 'TAX' and 'RAD' is visualized. The more the TAX the more the RAD is.  
- Corelation of 0.76 between 'INDUS' and 'NOX' is visualized.
- Another strong corelation of 0.73 between 'AGE' and 'NOX' is visualized.
- Another strong corelation of 0.73 between 'TAX' and 'INDUS' is visualized.

#### Negative corelation
- 'LSTAT' and 'MEDV' are negatively corelated with eachother
- same for 'PTRATIO' vs 'RM'

#### Pair plot to visualize impactful features

```python
key_features = ["RM", "LSTAT", "PTRATIO", "TAX", "MEDV", "RAD", "INDUS", "NOX", "AGE"]
sns.pairplot(df[key_features])
plt.show()
```

```python
#Scatter plots help to see how each feature correlates with MEDV.

fig, axes = plt.subplots(3, 2, figsize=(12, 10))

features = ["RM", "LSTAT", "TAX", "PTRATIO", "NOX", "AGE"]
for i, feature in enumerate(features):
    row, col = divmod(i, 2)
    sns.scatterplot(x=df[feature], y=df["MEDV"], ax=axes[row, col])
    axes[row, col].set_title(f"{feature} vs MEDV")

plt.tight_layout()
plt.show()
```

#### Observation:
- RM vs MEDV: Positive correlation—more rooms tend to increase house prices.

- LSTAT vs MEDV: Negative correlation—higher % lower status population is associated with lower prices.

- TAX vs MEDV: Slight negative trend—higher tax rates may correlate with lower prices.

- PTRATIO vs MEDV: Weak negative correlation—higher student-teacher ratios may lead to lower prices.

- NOX vs MEDV: Negative correlation—higher pollution levels may reduce house prices.

- AGE vs MEDV: Slight negative correlation—older houses tend to have lower prices.

#### Observation: 
- There are more property that has low to mid range in price, i.e. 2million to 8million
- Insignificant amount of counts that has higher price. 

```python
# Box plot for LSAT and RM with respect to MEDV


# Boxplot for LSTAT

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=pd.qcut(df["LSTAT"], q=4, labels=["Low", "Medium-Low", "Medium-High", "High"]), y=df["MEDV"])
plt.xlabel("LSTAT (Low to High)")
plt.ylabel("MEDV")
plt.title("House Prices vs LSTAT")
plt.tight_layout()

plt.show()
```

```python
# Boxplot for RM
plt.subplot(1, 2, 2)
sns.boxplot(x=pd.qcut(df["RM"], q=4, labels=["Low", "Medium-Low", "Medium-High", "High"]), y=df["MEDV"])
plt.xlabel("RM (Low to High)")
plt.ylabel("MEDV")
plt.title("House Prices vs RM")

plt.tight_layout()
plt.show()
```

LSTAT vs MEDV:

- Higher LSTAT (indicating lower-income areas) corresponds to lower median house prices.

-  The spread in house prices is wider in lower LSTAT groups, suggesting greater price variation in wealthier areas.

- There are a few outliers in high LSTAT areas where house prices are unexpectedly high.

RM vs MEDV:

- More rooms (higher RM) are associated with higher house prices.

- Houses with fewer rooms (low RM) have a much smaller price range, indicating they are generally lower in value.

- Some high-RM houses have significantly higher prices, possibly luxury homes.

These insights confirm that RM (number of rooms) positively impacts house prices, while LSTAT (low-income percentage) negatively impacts them.

#### Outliers

```python
# Outliers information using IQR method 

outlier_info = {}

for cols in df.columns:
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[cols] < lower_bound) | (df[cols] > upper_bound)]
    outlier_info[cols] = len(outliers)


print("Outlier Count per Feature:\n", outlier_info)
```

#### Observation
- No null and missing values found. 
- There are some outliers. Counts are displayed in the above cell. However, they don't play significant role to hinder the model.    Thus represented.
